```python
years = list(range(2023, 2021, -1))
```

```python
all_matches = []
```

```python
initial_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
import time
for year in years:
    data = requests.get(initial_url)
    soup = BeautifulSoup(data.text)
    team_table = soup.select('table.stats_table')[0]
    
    href = [l.get("href") for l in team_table.find_all('a')]
    links = [l for l in href if '/squads/' in l]
    team_links = [f"https://fbref.com{l}" for l in links]
    
    previous_season = soup.select("a.prev")[0].get("href")
    initial_url = f"https://fbref.com/{previous_season}"
    
    for team_link in team_links:
        team_name = team_link.split("/")[-1].replace("-Stats", "").replace("-", " ")
        
        data = requests.get(team_link)
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        
        soup = BeautifulSoup(data.text)
        links = soup.find_all('a')
        links = [l.get("href") for l in links]
        links = [l for l in links if l and '/all_comps/shooting' in l]
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt", "xG", "G-xG"]], on="Date")
        except ValueError:
            continue
        
        team_data = team_data[team_data["Comp"] == "Premier League"]
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(10)
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```

```python
match_df.to_csv("matches.csv")
```
