```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv", index_col=0)
```

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opponent_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["referee_code"] = matches["referee"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["result_code"] = matches["result"].astype("category").cat.codes
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
matches[matches["team"] == "Arsenal"]
```

```python
train = matches[matches["date"] < '2022-06-01']
```

```python
test = matches[matches["date"] > '2022-06-01']
```

```python
predictors = ["venue_code", "opponent_code", "day_code", "referee_code", "hour"]
```

```python
rf.fit(train[predictors], train["result_code"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import precision_score
```

```python
acc = accuracy_score(test["result_code"], preds)
```

```python
grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Arsenal").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    # Closed not working
    rolling_stats = group[cols].rolling(3).mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt", "xg_x", "g-xg", "poss"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling.droplevel('team')
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-06-01']
    test = data[data["date"] > '2022-06-01']
    rf.fit(train[predictors], train["result_code"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["result_code"], prediction=preds), index=test.index)
    acc = accuracy_score(test["result_code"], preds)
    return combined, acc
```

```python
combined, acc = make_predictions(matches_rolling, predictors + new_cols)
```
